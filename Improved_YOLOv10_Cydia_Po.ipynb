```
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
from time import sleep
from time import sleep
import torch

# Define the paths
Path1 = "/content/gdrive/MyDrive/Colab Notebooks/Cydia_Zarboubi_Article_2/Improved_YOLOv10/detect/img11.png"
ModelPath = "/content/gdrive/MyDrive/Colab Notebooks/Cydia_Zarboubi_Article_2/Improved_YOLOv10/ultralytics/runs/detect/train/weights/best.pt"
ImagePath = "/content/gdrive/MyDrive/Colab Notebooks/Cydia_Zarboubi_Article_2/Database/Cydia Test/img1.png"

# Load the model
model = YOLO(ModelPath)

# Predict on an image with a confidence threshold of 0.50
results = model.predict(ImagePath, conf=0.50)

# Custom plotting function
def plot_results(results, save_path, text_size=1.0, box_color=(0, 0, 255), text_bg_color=(255, 255, 255)):
    img = results[0].orig_img  # Original image

    # Ensure tensors are on the CPU
    boxes = results[0].boxes.xyxy.cpu().numpy() if torch.is_tensor(results[0].boxes.xyxy) else results[0].boxes.xyxy
    confs = results[0].boxes.conf.cpu().numpy() if torch.is_tensor(results[0].boxes.conf) else results[0].boxes.conf
    classes = results[0].boxes.cls.cpu().numpy() if torch.is_tensor(results[0].boxes.cls) else results[0].boxes.cls

    for box, conf, cls in zip(boxes, confs, classes):
        x1, y1, x2, y2 = map(int, box)
        label = f'{results[0].names[int(cls)]} {conf:.2f}'

        # Draw rectangle
        cv2.rectangle(img, (x1, y1), (x2, y2), box_color, 4)

        # Draw label with specified font size
        font = cv2.FONT_HERSHEY_SIMPLEX
        scale = text_size  # Font scale
        thickness = 2  # Thickness of the lines used to draw the text
        size = cv2.getTextSize(label, font, scale, thickness)[0]

        # Draw background rectangle for the text
        cv2.rectangle(img, (x1, y1 - size[1] - 13), (x1 + size[0], y1), box_color, -1)

        # Draw text
        cv2.putText(img, label, (x1, y1 - 10), font, scale, text_bg_color, thickness)

    # Count the number of occurrences for each class
    class_name_count_0 = 'Cydia_Po'
    l_0 = 0
    counts = {}
    for result in results:
        boxes = result.boxes.cpu().numpy()
        for box in boxes:
            cls = int(box.cls[0])
            if cls not in counts:
                counts[cls] = 1
            else:
                counts[cls] += 1

    for key in counts.keys():
        if class_name_count_0 == results[0].names[key]:
            l_0 = counts[key]

    # Save the number of pests
    numberp1 = l_0

    # Display the count on the image
    position_0 = (20, 40)
    txt_0 = f"Number of {class_name_count_0} = {l_0}"
    cv2.putText(img, txt_0, position_0, cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0, 255), 3)

    # Save the annotated image
    cv2.imwrite(save_path, img)

# Plot the results with custom text size, bounding box color, and text background color, and save to Path1
plot_results(results, Path1, text_size=1, box_color=(0, 0, 255), text_bg_color=(255, 255, 255))

# Read and display the saved image using cv2_imshow
img1 = cv2.imread(Path1)
cv2_imshow(img1)
```

```
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Cydia_Zarboubi_Article_2/Improved_YOLOv10/ultralytics/runs/detect/train/results.csv')

# Define the metrics columns to display
metrics = [
    "train/box_loss", "train/cls_loss", "train/dfl_loss",
    "metrics/precision(B)", "metrics/recall(B)",
    "val/box_loss", "val/cls_loss", "val/dfl_loss",
    "metrics/mAP50(B)", "metrics/mAP50-95(B)"
]

# Set the figure size and the number of subplots
fig, axes = plt.subplots(2, 5, figsize=(18, 8))
axes = axes.flatten()

# Loop to create each subplot
for i, metric in enumerate(metrics):
    if metric in df.columns:
        axes[i].plot(df.index, df[metric], marker='o', label='results')  # Plot the actual values
        axes[i].plot(df.index, df[metric].rolling(window=5).mean(), linestyle='--', label='smooth')  # Smoothed line
        axes[i].set_title(metric)
        axes[i].legend()

# Adjust the spacing between subplots
plt.tight_layout()

# Save the figure in EPS format
plt.savefig("/content/gdrive/MyDrive/Colab Notebooks/Cydia_Zarboubi_Article_2/Improved_YOLOv10/ultralytics/runs/detect/train/results_plot.eps", format='eps', dpi=300)

plt.show()
```

# **Calculer le nombre de FLOPs (Floating Point Operations) pour un modle YOLO**

```
%pip install fvcore
%pip install ultralytics
import ultralytics
ultralytics.checks()
```

```
from ultralytics import YOLO
import torch
from fvcore.nn import FlopCountAnalysis, flop_count_table

# Define the paths
ModelPath = "/content/gdrive/MyDrive/Colab Notebooks/Cydia_Zarboubi_Article_2/Improved_YOLOv10/ultralytics/runs/detect/train/weights/best.pt"

# Load the model
model = YOLO(ModelPath)

# Define a dummy input tensor (adjust dimensions as per your YOLO model)
dummy_input = torch.randn(1, 3, 640, 640).to('cuda')  # Example for an input size of 640x640

# Move the model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.model.to(device)

# Calculate FLOPs using fvcore
flop_analyzer = FlopCountAnalysis(model.model, dummy_input)
flops = flop_analyzer.total()
flops_in_billion = flops / 1e9  # Convert to billions

# Display the FLOPs details
print("")
print(f"Estimated FLOPs: {flops_in_billion:.2f} Billion")
#print(flop_count_table(flop_analyzer))
```
